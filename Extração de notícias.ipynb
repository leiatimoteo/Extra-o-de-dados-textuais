```python
import requests
from bs4 import BeautifulSoup
```

```python
# URL da vítima (pegar osdados da primeira página)
url = "https://olhardigital.com.br/editorias/noticias/"

# Cabeçalho para simular um navegador para não ser banido do site
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36"
}
```

```python
response = requests.get(url, headers=headers)
response
#requests é da biblioteca
#get é para acessar o site
#url é o url do site
# <Response [200]> -> requisição foi bem-sucedida
```

```python
soup = BeautifulSoup(response.text, "html.parser")
# pegar o texto do html
```

```python
soup
```

```python
# Exibe o html como texto
print(soup.get_text())
```

```python
#estamos interessados nas noticias, preciso pegar os links
```

```python
# Localizando o bloco com o id "textoMateria"
bloco_noticias = soup.find("div", class_="sidebar-container")

bloco_noticias
```

```python
links = [a['href'] for a in bloco_noticias.find_all('a', href=True)]
        
# Exibindo os links extraídos
for link in links:
    print(link)
```

```python
#retirar os links das próximas páginas olhando o padrão das noticias
links = [link for link in links if "https://olhardigital.com.br/2024" in link]

for link in links:
    print(link)
```

```python
#acessando as notícias
full_url =  links[0]  # Construindo a URL completa
full_url
response = requests.get(full_url, headers=headers)
full_url

```

```python
soup = BeautifulSoup(response.text, "html.parser")
soup
```

```python
print(soup.get_text(separator="\n").strip())
```

```python
# Extraindo o título
title = soup.find("meta", property="og:title")["content"]

# Extraindo a descrição
description = soup.find("meta", attrs={"name": "description"})["content"]

# Extraindo as indexações (tags)
tags = [tag["post-excerpt"] for tag in soup.find_all("meta", attrs={"name": "sf_indexacao"})]

date = soup.find("meta", property="article:modified_time")["content"]

# Exibindo os resultados
print("Título:", title)
print("Descrição:", description)
print("Indexações:", tags)
print("Data:", date)
```

```python
# Localizando o bloco de texto
texto_materia = soup.find("div", class_="post-content wp-embed-responsive")
texto_materia


# Extraindo e limpando o texto
texto = texto_materia.get_text(separator="\n").strip()
texto
```

```python
import time
import random
```

```python
informacoes = {
    "titulo": title,
    "descricao": description,
    "texto": texto,
    "data": date,
}
informacoes
```

```python
numero_paginas = 20

links_noticias = []

for p in range(numero_paginas):
    url = f"https://olhardigital.com.br/editorias/noticias/page/{p}"
    response = requests.get(url, headers=headers)

    print(p, response)
    
    soup = BeautifulSoup(response.text, "html.parser")
    bloco_noticias = soup.find("div", class_="sidebar-container")

    links = [a['href'] for a in bloco_noticias.find_all('a', href=True)]
    links = [link for link in links if "https://olhardigital.com.br/2024" in link]

    links_noticias += links

    # Esperar alguns segundos antes de ir para a próxima página
    tempo_espera = random.uniform(0, 0.01)
    time.sleep(tempo_espera)
```

```python
len(links_noticias)
links_noticias
```

```python
import json
import pathlib

data_dir = pathlib.Path("data")  # Diretório onde os arquivos serão salvos
data_dir.mkdir(exist_ok=True) 

for i, link in enumerate(links_noticias):
    full_url =  link  # Construindo a URL completa
    response = requests.get(full_url, headers=headers)

    soup = BeautifulSoup(response.text, "html.parser")

    title = soup.find("meta", property="og:title")["content"]
    description = soup.find("meta", attrs={"name": "description"})["content"]

    
    texto_materia = soup.find("div", class_="post-content wp-embed-responsive")
    texto = texto_materia.get_text(separator="\n").strip()

    informacoes = {
        "url": full_url,
        "titulo": title,
        "descricao": description,
        "texto": texto,
    }
    
    filename = f"noticia_{i}.json"
    
    file_path = data_dir / filename

    with file_path.open("w", encoding="utf-8") as f:
        json.dump(informacoes, f, ensure_ascii=False, indent=4)
    
    print(f"Arquivo salvo: {file_path}")
    # tempo_espera = random.uniform(0, .5)
    time.sleep(tempo_espera)  
```
